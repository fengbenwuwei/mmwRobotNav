```python
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

```python
csv_file_name='Bowlus_Tx_4.csv'
df = pd.read_csv(csv_file_name)
df = df[['1_aoa_az', 'rx_pos_ind_row', 'rx_pos_ind_column']]
df['rxPosInd_1'] = (df['rx_pos_ind_row'])*3-2
df['rxPosInd_2'] = (df['rx_pos_ind_column'])*3-2
angle = np.empty((480,480))
angle[:, :] = np.NaN

for index, row in df.iterrows():
    angle[int(row['rxPosInd_1']-1), int(row['rxPosInd_2']-1)] = row['1_aoa_az']
    angle[int(row['rxPosInd_1']-1), int(row['rxPosInd_2'])] = row['1_aoa_az']
    angle[int(row['rxPosInd_1']-1), int(row['rxPosInd_2']+1)] = row['1_aoa_az']

    angle[int(row['rxPosInd_1']), int(row['rxPosInd_2']-1)] = row['1_aoa_az']
    angle[int(row['rxPosInd_1']), int(row['rxPosInd_2'])] = row['1_aoa_az']
    angle[int(row['rxPosInd_1']), int(row['rxPosInd_2']+1)] = row['1_aoa_az']

    angle[int(row['rxPosInd_1']+1), int(row['rxPosInd_2']-1)] = row['1_aoa_az']
    angle[int(row['rxPosInd_1']+1), int(row['rxPosInd_2'])] = row['1_aoa_az']
    angle[int(row['rxPosInd_1']+1), int(row['rxPosInd_2']+1)] = row['1_aoa_az']

plt.imshow(angle)
plt.colorbar()
plt.savefig('ray_tracing_strongest_aoa.png', dpi=500)
```
